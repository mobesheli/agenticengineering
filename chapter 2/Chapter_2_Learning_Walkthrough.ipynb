## Start Here: two ways to use this chapter

The author is giving you two valid ways to learn this project. Choose the one that matches how you like to work.

### Option 1: learn with the notebook

Use this notebook if you want the chapter explained step by step. It shows the commands, the data, the agent structure, and the workflow in teaching order.

### Option 2: run the project without the notebook

Use the terminal only if you want to run the app directly. You can complete the whole chapter example with `main.py` and `pytest` without opening Jupyter at all.

If you are not sure which path to take, start with the notebook.

The run order below is the simplest first path for both options.

### One-time setup

```bash
cd "chapter 2"
python3.11 -m venv .venv
source .venv/bin/activate
python -m pip install -r requirements.txt
```

### Optional: open the notebook in Jupyter

```bash
python -m pip install notebook
jupyter notebook
```

### Best first run order

1. `python main.py describe`
   This shows the agent, tool, guardrail, and handoff setup.
2. `python main.py offline --claim CLM-1001`
   This runs the project without spending API tokens.
3. `python main.py offline --claim CLM-1002`
   This shows the missing-document path.
4. `python main.py offline --claim CLM-1003`
   This shows the escalation path.
5. `pytest`
   This confirms the sample project still works.
6. `python main.py live --claim CLM-1001`
   Only do this after setting `OPENAI_API_KEY`.


````python
from pathlib import Path
import json
import os
import subprocess
import sys

from claims_pipeline.agents import (
    build_claims_intake_agent,
    build_claims_reviewer_agent,
    build_router_agent,
)
from claims_pipeline.data import build_claim_input, get_claim, list_claim_ids
from claims_pipeline.guardrails import is_claim_related, validate_claim_verdict
from claims_pipeline.pipeline import run_offline_pipeline

chapter_root = Path.cwd()
if not (chapter_root / "main.py").exists():
    raise RuntimeError(
        "Open this notebook from the chapter 2 folder. Example: cd 'chapter 2' && jupyter notebook"
    )

def show_json(value):
    if hasattr(value, "model_dump"):
        value = value.model_dump()
    print(json.dumps(value, indent=2))

print(f"Working directory: {chapter_root}")
print("Available claims:", ", ".join(list_claim_ids()))
````

## 1. Understanding the command options

The CLI is intentionally small. Each command has one teaching purpose.

| Command | When to use it | What it shows |
| --- | --- | --- |
| `python main.py list-claims` | when you want sample claim ids | the three built-in claims |
| `python main.py describe` | when you want to understand the app structure | agents, tools, guardrails, and handoffs |
| `python main.py offline --claim CLM-1001` | when you want the safest first run | deterministic local workflow |
| `python main.py live --claim CLM-1001` | when you are ready for a real model call | OpenAI Agents SDK run |
| `pytest` | when you want to verify the sample app | automated checks |

The next cell shows the built-in help text first, then the `describe` output. This is the best starting point for a new reader.


````python
subprocess.run([sys.executable, "main.py", "-h"], cwd=chapter_root, check=False)
print("\n--- describe ---\n")
subprocess.run([sys.executable, "main.py", "describe"], cwd=chapter_root, check=False)
````

## 2. Looking at the sample claims

The project uses three small claims so readers can see three different outcomes:

- a straightforward approval
- a missing-document request
- a human-escalation case

That makes the workflow easy to learn without a large dataset.


````python
for claim_id in list_claim_ids():
    claim = get_claim(claim_id)
    print(f"{claim.claim_id} | incident={claim.incident_type} | amount={claim.amount}")
    print(f"  documents: {', '.join(claim.submitted_documents)}")
    print(f"  summary: {claim.summary}\n")
````

## 3. Seeing the exact input that enters the workflow

Before looking at the model-facing code, look at the exact claim text that gets passed into the pipeline. This makes the rest of the chapter much easier to follow.


````python
sample_claim = get_claim("CLM-1001")
print(build_claim_input(sample_claim))
````

## 4. Understanding guardrails and deterministic rules

Chapter 2 is not only about calling a model. It is also about drawing boundaries around the model.

This sample app does that in two simple ways:

- input guardrail: reject text that does not look like an insurance claim
- output rule: require citations and valid confidence
- deterministic rule: force human review for high-risk claims even if the model sounds confident


````python
print("Relevant claim text:", is_claim_related("Claim CLM-1001 for policy POL-1001 needs review."))
print("Unrelated text:", is_claim_related("Summarize this restaurant menu."))
print("\nOutput validation for an offline verdict:")
verdict = run_offline_pipeline("CLM-1001").verdict
print(validate_claim_verdict(verdict))
````

## 5. Inspecting the agents, tools, and handoffs

This is the direct map from the chapter language to the codebase:

- intake agent: checks completeness
- reviewer agent: uses tools and returns a structured verdict
- router agent: demonstrates handoffs to specialists


````python
intake = build_claims_intake_agent()
reviewer = build_claims_reviewer_agent()
router = build_router_agent()

show_json(
    {
        "intake_agent": {
            "name": intake.name,
            "tools": [tool.name for tool in intake.tools],
            "input_guardrails": [guardrail.name for guardrail in intake.input_guardrails],
            "output_type": intake.output_type.__name__,
        },
        "reviewer_agent": {
            "name": reviewer.name,
            "tools": [tool.name for tool in reviewer.tools],
            "output_guardrails": [guardrail.name for guardrail in reviewer.output_guardrails],
            "output_type": reviewer.output_type.__name__,
        },
        "router_agent": {
            "name": router.name,
            "handoffs": [agent.name for agent in router.handoffs],
        },
    }
)
````

## 6. Running the offline learning path

This is the best way to teach the project because the results are stable every time.

- `CLM-1001` should approve
- `CLM-1002` should request documents
- `CLM-1003` should escalate

If these three outcomes make sense, the reader understands the core workflow.


````python
for claim_id in list_claim_ids():
    result = run_offline_pipeline(claim_id)
    print(f"\n=== {claim_id} -> {result.stage} ===")
    show_json(result)
````

## 8. Running the live SDK path

Only use the live path after the offline flow is clear.

### Exact commands

```bash

python main.py live --claim CLM-1001
```

Use `CLM-1001` first because it is the simplest case. After that, try `CLM-1002` and `CLM-1003` and compare the live output to the offline result in this notebook.

If the key is valid but the project has no quota, the CLI exits with a clear message.

The next cell performs one real live SDK run for `CLM-1001` when `OPENAI_API_KEY` is set.


````python
print("OPENAI_API_KEY configured:", bool(os.getenv("OPENAI_API_KEY")))
if os.getenv("OPENAI_API_KEY"):
    print("\n--- live CLM-1001 ---\n")
    subprocess.run(
        [sys.executable, "main.py", "live", "--claim", "CLM-1001"],
        cwd=chapter_root,
        check=False,
    )
else:
    print("Set OPENAI_API_KEY and rerun this cell.")
    print("Recommended first live command: python main.py live --claim CLM-1001")
````
